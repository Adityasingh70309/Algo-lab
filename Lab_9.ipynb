# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance

    return distances
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```

```python
def d(g, s):
    ds = {n: float('inf') for n in g}
    ps = {n: None for n in g}
    ds[s] = 0
    q = list(g.keys())

    while q:
        u = min(q, key=lambda n: ds[n])
        q.remove(u)

        if ds[u] == float('inf'):
            break

        for v, w in g[u].items():
            a = ds[u] + w
            if a < ds[v]:
                ds[v] = a
                ps[v] = u
    return ds, ps

def p(ps, s, t):
    path = []
    c = t
    while c is not None and c != s:
        path.insert(0, c)
        c = ps[c]
    if c == s:
        path.insert(0, s)
        return path
    return []

def g_inp():
    gr = {}
    while True:
        node = input("Enter node name (or 'done' to finish nodes): ")
        if node == 'done':
            break
        gr[node] = {}
        while True:
            n_inp = input(f"Enter neighbor and weight for {node} (e.g., B 4, or 'done' to finish neighbors for {node}): ")
            if n_inp == 'done':
                break
            try:
                n, w = n_inp.split()
                gr[node][n] = int(w)
            except ValueError:
                print("Invalid input. Use 'NODE WEIGHT' format.")
    return gr

g = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 5, 'D': 10, 'E': 6},
    'C': {'A': 2, 'B': 5, 'D': 3, 'E': 8},
    'D': {'B': 10, 'C': 3, 'F': 7},
    'E': {'B': 6, 'C': 8, 'F': 1},
    'F': {'D': 7, 'E': 1}
}

print("Graph used:")
for n, edges in g.items():
    print(f"  {n}: {edges}")
print("-" * 20)

srcs = ['A', 'B', 'E']

for src in srcs:
    if src not in g:
        print(f"Source node {src} not in graph.")
        print("-" * 20)
        continue

    print(f"Shortest paths from source: {src}")
    ds, ps = d(g, src)

    for dest in sorted(g.keys()):
        path_nodes = p(ps, src, dest)
        if path_nodes:
            path_str = " -> ".join(path_nodes)
            print(f"  Path from {src} to {dest}: {path_str}, Total distance: {ds[dest]}")
        else:
            print(f"  Path from {src} to {dest}: No path found, Total distance: {ds[dest]}")
    print("-" * 20)
```
